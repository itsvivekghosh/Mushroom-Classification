```python
cols=X.columns
len(cols)
```

```python
for col in cols:
    
    plt.xlabel("Class")
    plt.ylabel(str(col)+" Category")
    plt.title("Mushrooms edible and poisonous in {} category".format(col))
    sns.countplot(x=df['class'], hue=df[col])
    plt.legend(loc='right')
    plt.show()
    
    print("In Edible Category, {} category has maximum mushrooms in {} ie. {}".format(df[df['class'] == 'e'][col].value_counts().index[0], col, df[df['class'] == 'e'][col].value_counts().max()))
    print("In Poisonous Category, {} category has maximum mushrooms in {} ie. {}".format(df[df['class'] == 'p'][col].value_counts().index[0], col, df[df['class'] == 'p'][col].value_counts().max()))
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le
```

```python
for col in cols:
    
    X[col]=le.fit_transform(X[col])
    
X.head()
y=pd.DataFrame(le.fit_transform(y), columns=['class'])
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc
```

```python
X=pd.DataFrame(sc.fit_transform(X), columns=[cols])
X.shape
```

```python
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.model_selection import train_test_split as tts
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
X_train, X_test, y_train, y_test=tts(X, y, test_size=0.33, random_state=0)
print(X_train.shape, X_test.shape)
```

```python
rf=RandomForestClassifier()
rf
```

```python
rf.fit(X_train, y_train)
accuracies=cross_val_score(estimator=rf, X=X_test, y=y_test, n_jobs=-1, cv=5, scoring='r2')
accuracies.mean()
```

```python
y_pred=rf.predict(X_test)
y_pred[:10], y_test.head(10)
```

```python
sns.heatmap(confusion_matrix(y_pred, y_test), annot=True)
```

```python
print("Accuracy Score: {}".format(accuracy_score(y_test, y_pred)))
print("Classification Report:\n {}\n".format(
    classification_report(y_test, y_pred)
))
```
